```python
import matplotlib.pyplot as plt
```

```python
xval=[1,2,3,4,5,6]
yval1=[x**2 for x in xval]
yval2=[x**3 for x in xval]
yval3=[x**4 for x in xval]
plt.title("Exponent Analysis",fontsize=30)
plt.xlabel("Numbers",fontsize=15)
plt.ylabel("Powers upto 4", fontsize=15)
plt.tick_params(axis="both", labelsize=9)
plt.scatter(xval,yval1,c="red",s=70)
plt.plot(xval,yval1,linewidth=1,c=(0.02,0.7,0.32))
plt.scatter(xval,yval2,s=40,c=(0.4,0.4,0.5))
plt.plot(xval,yval2,linewidth=1,c=(0.02,0.4,0.5))
plt.scatter(xval,yval3,c=yval3,cmap=plt.cm.Blues)
plt.show()
plt.savefig('analysis.png')
```
