```python
from random import randint
```

```python
class Dice():
    def __init__(self,maxnumber=6):
        self.maxnumber=maxnumber
    def diceval(self):
        return randint(1,self.maxnumber)
    
```

```python
die=Dice()
results=[]
for i in range(15):
    result=die.diceval()
    results.append(result)
print(results)
frequencies=[]
for j in range(1,die.maxnumber+1):
    frequency=results.count(j)
    frequencies.append(frequency)
print(frequencies)
```

```python
import pygal
hist=pygal.Bar()
hist.add("D6",frequencies)
hist.render_to_file("bargraph.svg")
```
