```python
import matplotlib.pyplot as plt
from random import randint
class Die():
    def __init__ (self,maxnum=6):
        self.maxnum=maxnum
    def rollval(self):
        return randint(1,self.maxnum)
die=Die()
values = []
for x in range(20):
    value=die.rollval()
    values.append(value)

frequencies = []
for y in range(1,die.maxnum+1):
    countval=values.count(y)
    frequencies.append(countval)
xval=[]
for z in range(1,21):
    xval.append(z)
xval2=[]
for a in range(1,die.maxnum+1):
    xval2.append(a) 
    
plt.title("Random Dice Roll Analysis",fontsize=20)
plt.xlabel("X values",fontsize=15)
plt.ylabel("Values (blue) and frequencies (red)",fontsize=10)
plt.tick_params(axis="both",labelsize=9);
plt.scatter(xval,values,s=70,c="blue")
plt.plot(xval,values,linewidth=2)
plt.scatter(xval2,frequencies,s=90,c="red")
plt.plot(xval2,frequencies,linewidth=2)
plt.show()
```
